## 5 класс  

Необходимо написать программу, которая рисует с помщью модуля `turtle` случайное количество случайных фигур.  
Система оценивания:  
* **на оценку 3** напишите как минимум три функции, рисующие различные геометрические фигуры. Например, треугольник, квадрат, прямоугольник.  
* **на оценку 4** Используя эти функции, нарисуйте несколько фигур, выбирая их случайным образом. (можно использовать функцию `turtle.circle`)
* **на оценку 5** Выбирайте размер и цвет фигур случайным образом, передавая их как аргумент в функцию.
